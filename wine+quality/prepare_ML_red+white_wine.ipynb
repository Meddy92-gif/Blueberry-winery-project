confusion matrix for RFC in red wine data

```python
from sklearn.metrics import confusion_matrix

print("Confusion matrix:")
print(confusion_matrix(y_test_red, y_pred_rwq ))
```

This confusion matrix represents the performance of a Random Forest Classifier (RFC) on predicting the quality of red wine. The rows represent the actual (true) labels, while the columns represent the predicted labels.

#### Breakdown:
#### True label "High" (first row):

108 wines were correctly classified as "High".
33 wines were incorrectly classified as "Medium".
0 wines were incorrectly classified as "Low".

#### True label "Medium" (second row):

36 wines were incorrectly classified as "High".
137 wines were correctly classified as "Medium".
1 wine was incorrectly classified as "Low".

#### True label "Low" (third row):

0 wines were incorrectly classified as "High".
5 wines were incorrectly classified as "Medium".
0 wines were correctly classified as "Low".
Key Takeaways:
The classifier performed well in classifying "Medium" wines (137 correct out of 174).
There are some misclassifications between "High" and "Medium" wines.
The classifier struggled with "Low" wines (all 5 were misclassified as "Medium").

```python
cm_red = confusion_matrix(y_test_red, y_pred_rwq)

# Define class labels
labels = ["High", "Medium", "Low"]

# Create a heatmap
plt.figure(figsize=(6,5))
sns.heatmap(cm_red, annot=True, fmt="d", cmap="Reds", xticklabels=labels, yticklabels=labels)

# Add labels and title
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion Matrix for Red wine - Random Forest Classifier (RLC)")

# Show the plot
plt.show()
```

splitting red wine data for ML model

```python
from sklearn.model_selection import train_test_split, GridSearchCV
winequality_ml = red_wine.copy()

# split data into X and Y
X_red = red_wine.drop(['fixed acidity','pH','chlorides','quality_label','total sulfur dioxide','residual sugar','free sulfur dioxide','density','quality'], axis=1)
Y_red = red_wine['quality_label']

# split data
x_train_red, x_test_red, y_train_red, y_test_red = train_test_split(X_red, Y_red, test_size=0.2, random_state=1)
```

training the ML model (Decisison Tree Classifier - DTC) on Red Wine Data

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree


clf = DecisionTreeClassifier()
clf = clf.fit(x_train_red, y_train_red)
plot_tree(clf, filled=True)
plt.title("Decision tree trained on Red_Wine Data")
plt.show()
```

Testing the DTC Model on Red Wine (RW) Data

```python
y_pred_clf_rw = clf.predict(x_test_red)

from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test_red, y_pred_clf_rw)
print(acc)

# Model Accuracy, how often is the classifier correct?

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test_red, y_pred_clf_rw))
```
