```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define your data
data_x = ['LOR_model', 'RFC', 'DTC', 'SVM']
data_y = [0.67, 0.72, 0.64, 0.72]

# Create a DataFrame
df_ML = pd.DataFrame({
    'Model': data_x,
    'Accuracy': data_y
})

# Create a bar plot using seaborn
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] 
sns.barplot(x='Model', y='Accuracy', width=0.5, data=df_ML, palette=colors)
plt.title('Model Accuracy Comparison for Red Wine')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.show()
```
